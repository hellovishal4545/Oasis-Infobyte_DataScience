**Data Cleaning**

```python
#change categorical data into numerical value
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
df.head(1)
```

```python
df['Species'].unique()
```

```python
X = df.iloc[:,0:4]
X.head(1)
```

```python
y = df.iloc[:,-1]
y.head()
```

```python
print(X.shape)
print(y.shape)
```

**Model Building**

**Split data into testing & training**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
#confusion metrics
from sklearn.metrics import confusion_matrix ,accuracy_score,classification_report
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
acc = accuracy_score(y_test,y_pred)
print('Accuracy Score:',acc)
```

```python
print("Classification Report:")
print(classification_report(y_test,y_pred))
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)

cm1 = confusion_matrix(y_test,y_pred)
print(cm1)

acc1 = accuracy_score(y_test,y_pred)
print('Accuracy Score:',acc1*100)

print("Classification Report:")
print(classification_report(y_test,y_pred))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)

y_pred = dt.predict(X_test)

cm2 = confusion_matrix(y_test,y_pred)
print(cm2)

acc2 = accuracy_score(y_test,y_pred)
print('Accuracy Score:',acc2*100)

print('Classificatiin Report:')
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
#for visualizing the Decision Tree
feature = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
classes = ['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
plt.figure(figsize = (15,10))
plot_tree(dt,feature_names = feature,class_names = classes,filled = True)  #decision tree
plt.show()
```

**Support Vector Machinea**

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)

y_pred = svc.predict(X_test)

#confusion Matric
cm = confusion_matrix(y_test,y_pred)
print(cm)

acc = accuracy_score(y_test,y_pred)
print('Accuracy Score:',acc*100)

print("Classification Report:")
print(classification_report(y_test,y_pred))
```

**K-NN Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)

#confusion Matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)

#accuracy
acc = accuracy_score(y_test,y_pred)
print('Accuracy Score:',acc*100)

print('Classification Report:')
print(classification_report(y_test,y_pred))
```

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)

y_pred = nb.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)

acc = accuracy_score(y_test,y_pred)
print('Accuracy Score:',acc*100)

print('Classification Report:')
print(classification_report(y_test,y_pred))
```
